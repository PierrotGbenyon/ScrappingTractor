```python
import pandas as pd
```

Nettoyage du dataset

```python
data = pd.read_csv('tractors_dataset.csv')
```

```python
data.head()
```

Retirer les ":" au niveau des noms de colonnes

```python
data_cleaned = data.copy()
data_cleaned["Model"] = data_cleaned["Model:"]
data_cleaned["Condition"] = data_cleaned["Condition:"]
data_cleaned["Location"] = data_cleaned["Location:"]
data_cleaned["Listing ID"] = data_cleaned["Listing ID:"]
data_cleaned["Update"] = data_cleaned["Update:"]
data_cleaned["Year of construction"] = data_cleaned["Year of construction:"]
data_cleaned["Mileage"] = data_cleaned["Mileage:"]
# retirer les anciennes colonnes
data_cleaned.drop(columns=["Model:", "Condition:", "Location:", "Listing ID:", "Update:", "Year of construction:", "Mileage:"], inplace=True)
```

```python
data_cleaned.head()
```

Nettoyage de la colonne price

```python
data_cleaned["price"].unique()
```

```python
data_cleaned.shape
```

```python
prix_affiche = (data_cleaned['price'].str.replace('€', '').str.replace('$', '').str.strip().str.split(' ').str[0])
#
data_cleaned['prix_affiche'] = pd.to_numeric(prix_affiche.str.replace(',', ''), errors='coerce')

# Affichage pour vérifier
data_cleaned['prix_affiche'].head()
```

```python
data_cleaned.drop(columns=["price"], inplace=True)
```

```python
data_cleaned
```

```python
data_cleaned["location"].unique()
```

```python
data_cleaned["Location"].unique()
```

```python
data_cleaned["ville"] = data_cleaned["location"].astype(str)
data_cleaned.drop(columns=["location"], inplace=True)
data_cleaned["pays"] = data_cleaned["Location"].str.split(',').str[-1].str.strip()
data_cleaned.drop(columns=["Location"], inplace=True)   
```

```python
data_cleaned["pays"].unique()
```

```python
data_cleaned["pays"] = data_cleaned["pays"].str.replace('The Netherlands', 'Netherlands')
```

```python
data_cleaned["pays"].unique()
```

Gestion des valeurs manquantes

```python
data_cleaned.isnull().sum()
```

```python
mapping_villes_pays = {
    'Atlanta': 'United States',
    'Norway': None,           # "Norway" est un pays, aucune région norvégienne de votre liste ne correspond spécifiquement
    'Vuren': 'Netherlands',
    'Veghel': 'Netherlands',
    'Heteren': 'Netherlands',
    'Viersen': 'Germany',
    'Dortmund': 'Germany',
    'Madrid': 'Spain',
    'Sittensen': 'Germany',
    'Nordland': 'Nordland',   # Présent à la fois dans la liste des villes et des régions
    'Sagunto': 'Spain',
    'Lichtenfels': 'Germany',
    'Eilenburg': 'Germany'
}
```

```python
data_cleaned['pays'] = data_cleaned['pays'].fillna(data_cleaned['ville'].map(mapping_villes_pays))
```

```python
data_cleaned.isnull().sum()
```

```python
#convertir en entier
data_cleaned["Year of construction"] = data_cleaned["Year of construction"].astype('Int64')
#retirer km et convertir en float
data_cleaned["Mileage"] = data_cleaned["Mileage"].str.replace(' km', '').str.replace(',', '').astype('Int64')
```

```python
data_cleaned["Mileage"].unique()
```

```python
#data_cleaned["Mileage"].fillna(0, inplace=True)
```

```python
data_cleaned["Update"] = pd.to_datetime(data_cleaned['Update'], format='%d.%m.%Y', errors='coerce')
```
